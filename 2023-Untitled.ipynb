```python
import subprocess
import random
import datetime

# Replace with the actual path to the folder containing your .cpp files
path_to_folder = "/Users/shobhitsingh/Desktop/git/git_project"

# Specify the date range
start_date = datetime.date(2023, 1, 14)
end_date = datetime.date(2023, 3, 30)

# Get list of .cpp files
cpp_files = subprocess.check_output(f'ls {path_to_folder}/*.cpp', shell=True).decode().strip().split('\n')

# Generate and apply random dates and times for commits
for cpp_file in cpp_files:
    commit_date = start_date + datetime.timedelta(days=random.randint(0, (end_date - start_date).days))
    commit_time = datetime.time(random.randint(0, 23), random.randint(0, 59), random.randint(0, 59))
    commit_datetime = datetime.datetime.combine(commit_date, commit_time)
    
    commit_message = cpp_file.replace(".cpp", "")
    
    subprocess.call(f'cd {path_to_folder} && git add "{cpp_file}"', shell=True)
    subprocess.call(f'cd {path_to_folder} && GIT_COMMITTER_DATE="{commit_datetime}" git commit -m "{commit_message}"', shell=True)

subprocess.call(f'cd {path_to_folder} && git push DSA_origin master', shell=True)
```
